# 4. Filter the dataframe with only this columns

```python
df = df[['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index']]
```

```python
df
```

# 5. Data Cleaning

```python
# Remove all duplicates observations
df.drop_duplicates()
```

```python
df.duplicated().value_counts()
```

```python
# Find missing values in all columns
df.isnull()
```

```python
# Remove all observations where continent column value is missing.
df.dropna(subset=['continent'])
```

```python
# Fill all missing values with 0 
df.fillna(value=0)
```

# 6.Date time format

```python
df['date'] = pd.to_datetime(df['date'])
df
```

```python
# Create new column month after extracting month data from date column.
df['month'] = df['date'].dt.month
df
```

# 7. Data Aggregation

```python
# Find max value in all columns using groupby function on 'continent'.
df_groupby = df.groupby('continent').max().reset_index()
df_groupby
```

# 8. Feature Engineering

```python
# Create a new feature 'total_death_to_total_cases' by ratio of 
# 'total_deaths' columns to 'total_cases'.
df_groupby['total_deaths_to_total_cases'] = df_groupby['total_deaths']/df_groupby['total_cases']
```

```python
df_groupby
```

# 9.Data Visualization

```python
# Perform Univariate analysis on 'gdp_per_capita' column by plotting histogram using seaborn dist plot.
sns.displot(df['gdp_per_capita'],kde=False,color='blue')
```

```python
# Plot a scatter plot of 'total_cases' & 'gdp_per_capita'.
x = df_groupby['total_cases']
y = df_groupby['gdp_per_capita']
sns.scatterplot(x,y)
```

```python
# Plot Pairplot on df_groupby dataset.
sns.pairplot(df_groupby)
```
